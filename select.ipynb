```python
def mylistdir(dir):
    files=[]
    for file in os.listdir(dir):
        if file.endswith('.jpg'):
            files.append(file)
    
    return files

def myimread(name):
    pic=mpimg.imread(name)
    return np.dot(pic[...,:3], [0.299, 0.587, 0.114])

path=os.path.join('template')
files=mylistdir(path)
pic=0
for file in files:
    pic=pic+myimread(os.path.join(path,file))

pic=pic/7
plt.imsave(os.path.join('template','tem.jpg'),pic,cmap='gray')
```
